```python
import pandas as pd
```

```python
df = pd.read_csv('data.csv')
df
```

```python
df['ts'] = pd.to_datetime(df['ts'], errors = 'coerce')
df['hour'] = df['ts'].dt.hour
df['date'] = df['ts'].dt.date
df
```

```python
new_df = df[df['date'] == pd.to_datetime('2025-10-21').date()]
new_df
```

```python
result = (new_df
          .groupby('hour')['seconds']
          .sum()
          .sort_values(ascending=False)
          .head(5)
)
result
```
