```python
import pandas as pd
```

```python
df = pd.read_csv('data.csv')
df
```

```python
df['sentence1'] = df['sentence_1'].str.split(' ')
df['sentence2'] = df['sentence_2'].str.split(' ')
df[['sentence1','sentence2']]
```

```python
def find_common_words(row):
    return set(row['sentence1']) & set(row['sentence2'])
df['common_words'] = df.apply(find_common_words, axis = 1)
df['common_words']
```

```python
def count_common_words(row):
    result = {}

    for word in row['common_words']:
        count = row['sentence2'].count(word)
        result[word] = count
        
    return result
```

```python
df['final_output'] = df.apply(count_common_words, axis = 1)
df['final_output']
```
