```python
import csv

# All, Hollywood, Bollywood - Max, Min and AVG Ratings
data = []
with open('movies.csv','r') as movies_data:
    data = list(movies_data)
    header = data[0]
    print(header)
    data = data[1:]
    print(data)  
```

```python
def calculate_ratings(data, industry=None):
    ratings = []
    for row in data:
        row = row.strip().split(",")
        if row[3] != 'NULL' and (not industry or row[1] == industry):
            ratings.append(float(row[3]))
    max_rating = max(ratings)
    min_rating = min(ratings)
    avg_rating = sum(ratings) / len(ratings)
    return max_rating, min_rating, avg_rating
```

```python
max_rating, min_rating, avg_rating = calculate_ratings(data)
print(f'''All records: Min Rating = {min_rating}, Max Rating = {max_rating}, Average Rating = {avg_rating}''')
max_rating, min_rating, avg_rating = calculate_ratings(data, industry = 'Hollywood')
print(f'''Hollywood: Min Rating = {min_rating}, Max Rating = {max_rating}, Average Rating = {avg_rating}''')
max_rating, min_rating, avg_rating = calculate_ratings(data, industry = 'Bollywood')
print(f'''Bollywood: Min Rating = {min_rating}, Max Rating = {max_rating}, Average Rating = {avg_rating}''')
```
